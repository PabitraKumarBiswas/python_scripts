```python
import pandas as pd

# Define file path
excel_path = '/Users/pabitra/Documents/RIMES/ADSS/Bamis Data Share.xlsx'

# Read the sheet
sheet_name = 'Other Source - Diseases Informa'
df = pd.read_excel(excel_path, sheet_name=sheet_name)

# Detect Column B
host_col = df.columns[1]   # Column B

# Extract, clean, and deduplicate crop names
crop_list = (
    df[host_col]
    .dropna()
    .astype(str)
    .str.strip()        # remove leading/trailing spaces
    .str.title()        # normalize names for comparison
)

# Remove exact duplicates
crop_list = sorted(set(crop_list))

# Print result
print("\nðŸŒ¾ Unique crops in Column B:")
for crop in crop_list:
    print("-", crop)

print(f"\nâœ… Total unique crops found: {len(crop_list)}")
```
